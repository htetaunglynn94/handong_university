# Define functions

**Statements** $\to$ $f(x) = x^4+x^3-6x^2+4x+12$

### Given function

```python
# define given function
def given_func(x):
    y = pow(x,4) + pow(x,3) - 6*pow(x,2) + 4*x + 12 # x^4 + x^3 - 6x^2 + 4x + 12
    return y
```

### Derivative function

```python
# define derivative function
def derivative(func, x, h=1e-9):
    return (func(x + h) - func(x)) / h
```

### Plotting function

```python
# define function for visualization
def plot_graph(func, x_axis_start, x_asix_end):
    x = np.arange(x_axis_start, x_axis_end, 0.1) # set the range for x with the fixed interval 0.1
    print("number of points in X-axis:", len(x)) # print to know maximum number of values in X-axis
    plt.figure(figsize=(10,5)) # set figure size
    plt.plot(x, func(x), color='black') # plot x values vs y value (output of function) with black line
    plt.xlabel('X-axis') # label for X-axis
    plt.ylabel('Y-axis') # label for Y-xais
    plt.grid(color='gray', alpha=0.3, linestyle='--') # set grid color, transparency and line style
    return plt
```

```python
# Define func to implement Gradient Descent algorithm
def Gradient_Descent(func, x_initial, alpha, epsilon, max_number_iter):
    print("step_size:", alpha)
    print("initial point:", x_initial)
    count = 0 # variable for counting the number of iterations
    x_current = x_initial # set initial x value as current x
    plt = plot_graph(func, x_axis_start, x_axis_end)
    diff = derivative(func, x_current) # compute the derivative value at the current x
    plt.scatter(x_current, func(x_current), color='blue') # plot the initial x position with blue color
    while count < max_number_iter: # (2nd priority) terminate iterations when it exceeds the maximum number of
        x_next = x_current - alpha * diff
        count += 1
        plt.scatter(x_next, func(x_next), color='red', alpha=0.5) # plot next x position with red
        # if abs(x_current - x_next) < epsilon: # (1st priority) terminates if the difference is less than the priority
        #     break
        x_current = x_next # update the x position
        diff = derivative(func, x_current) # update the value of the derivative of the func at new x
    print('Number of iterations: {}\nOptimum(x,y) = ({},{})'.format(count, x_next, func(x_next)))
```
