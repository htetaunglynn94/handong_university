## Define Functions

### Create Function
**Statements** $\to$ $f(x) = x^4+x^3-6x^2+4x+12$

```python
def create_function(x):
    y = pow(x,4) + pow(x,3) - 6*pow(x,2) + 4*x + 12 # x^4 + x^3 - 6x^2 + 4x + 12
    return y
```

### Visualization Function

```python
# define function for visualization
def visualization(func, x_axis_start, x_axis_end):
    x = np.arange(x_axis_start, x_axis_end, 0.1) # set the range for x with the fixed interval 0.1
    plt.figure(figsize=(10,5)) # set figure size
    plt.plot(x, func(x), color='black') # plot x values vs y value (output of function) with black line
    plt.xlabel('X-axis') # label for X-axis
    plt.ylabel('Y-axis') # label for Y-xais
    plt.grid(color='gray', alpha=0.3, linestyle='--') # set grid color, transparency and line style
    return plt
```

### Derivative Function

```python
# define derivative function
def derivative(func, x, h=1e-9):
    return (func(x + h) - func(x)) / h # derivative formula referenced from ChatGPT
```

### Gradient Descent Function
Gradient Descent Formula $\to$ $x_k =x_{k - 1} - \alpha \nabla f(x_{k - 1})$

```python
# Define func to implement Gradient Descent algorithm
def Gradient_Descent(func, x_initial, learning_rate, max_iter):
    print("step_size (learning rate):", learning_rate) # print step size
    x = x_initial # set initial x value as current x
    plt = visualization(func, x_start, x_end) # plot function with black line
    plt.scatter(x, func(x), color='blue', alpha=0.5) # plot the initial position of point x with blue color
    
    for i in range(max_iter+1): # use for-loop for iterations
        diff_val = derivative(func, x) # compute the derivative value at the current x
        x -= learning_rate * diff_val # Gradient Descent formula (coding reference from Gemini)
        plt.scatter(x, func(x), color='red', alpha=0.7) # plot next x position with red color
    
    print(f'Number of iterations: {i}\nLocation of global minimum point = ({x},{func(x)})')
```
