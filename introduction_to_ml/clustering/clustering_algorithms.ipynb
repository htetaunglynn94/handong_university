# Clustering Algorithms

## Libraries

```python
# general libraries
import pandas as pd                               # for data manipulation
import numpy as np                                # for mathematical operations
import seaborn as sns                             # for data visualization
import matplotlib.cm as cm                         # to set matplotlib color map
import matplotlib.pyplot as plt                   # for data visualization

# sklearn libraries
from sklearn.cluster import KMeans                # Kmeans library
from sklearn.cluster import DBSCAN                #
from sklearn.datasets import make_blobs           #
from sklearn.datasets import make_moons           #
from sklearn.mixture import GaussianMixture       #
from sklearn.metrics import silhouette_score      #
from sklearn.metrics import silhouette_samples    #
```

```python
def plot_image(c, x, y):
    plt.scatter(x, y, c=c, marker='o',cmap='tab20',s=10)
    plt.xlabel('X1', fontsize=10)
    plt.ylabel('X2', fontsize=10)
    plt.grid(True, alpha=0.3, linestyle='--')
    # plt.savefig('Results_K-means_algorith', dpi=250, bbox_inches='tight')
    plt.show()
```

```python
def plot_sketch(x, y):
    plt.scatter(x, y, marker='o', color='black', s=10)
    plt.xlabel('X1', fontsize=10)
    plt.ylabel('X2', fontsize=10)
    plt.grid(True, alpha=0.3, linestyle='--')
    plt.show()
```

# K-Means algorithm

### Sketch with Data1

```python
# import "Data_1.csv"
data = pd.read_csv('Data_1.csv')
data.head()
```

```python
plot_sketch(x=data['x1'], y=data['x2'])
```

### Plot with Data1

```python
data_array = np.column_stack([data['x1'], data['x2']]) # stack 1D array to 2D
data_array[0:4]
```

```python
number_of_cluster = 3
kmeans_algorithm = KMeans(n_clusters=number_of_cluster, random_state=0).fit(data_array)
kmeans_labels = kmeans_algorithm.labels_
kmeans_cluster_center = kmeans_algorithm.cluster_centers_
```

```python
print(kmeans_labels[0:10])
print(kmeans_cluster_center)
```

```python
plot_image(kmeans_labels, data['x1'], data['x2'])
```

### With Data3

```python
# import "Data_3.csv"
data = pd.read_csv('Data_3.csv')
data.head()
```

```python
plot_sketch(x=data['x1'], y=data['x2'])
```

```python
data_array = np.column_stack([data['x1'], data['x2']]) # stack 1D array to 2D
data_array[0:4]

number_of_cluster = 2
kmeans_algorithm = KMeans(n_clusters=number_of_cluster, random_state=0).fit(data_array)
kmeans_labels = kmeans_algorithm.labels_
print(kmeans_labels)
kmeans_cluster_center = kmeans_algorithm.cluster_centers_
```

```python
plot_image(kmeans_labels, data['x1']*100, data['x2']*0.01)
```

```python
# Define hyper-parameters for the DBSCAN algorithm
dbscan_eps = 0.3
dbscan_min_sample = 1
```

```python
# Specify an input for the process of implementation
data_array = np.column_stack([data['x1'], data['x2']]) # stack 1D array to 2D
```

```python
# Implement the DBSCAN algorithm
from sklearn.cluster import DBSCAN
dbscan_algorithm = DBSCAN(eps=dbscan_eps, min_samples=dbscan_min_sample, metric='euclidean',
                         algorithm='ball_tree').fit(data_array)
dbscan_labels = dbscan_algorithm.labels_
```

```python
# Counter number of clusters in labels
dbscan_number_of_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
dbscan_number_of_clusters
```

```python
# Count number of noise points
dbscan_number_of_noise = list(dbscan_labels).count(-1)
dbscan_number_of_noise
```

```python
plot_image(dbscan_labels, data['x1'], data['x2'])
```
